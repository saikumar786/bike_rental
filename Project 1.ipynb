### Feature Scaling

```python
# In the columns we have selected,'casual' and 'registered' only not normalized
```

```python
#A checkpoint before Normalizing data
day_before_normalize = day.drop(['instant', 'dteday',], axis=1)
```

```python
day_before_normalize.head()
```

```python
#So, we Normalize those variables
cols_to_normalize = ['casual', 'registered']

for column in cols_to_normalize:
    day_before_normalize[column] = (day_before_normalize[column] - min(day_before_normalize[column]))/(max(day_before_normalize[column]) - min(day_before_normalize[column]))
```

```python
day_before_normalize.head()
```

```python
#Another checkpoint after Normalizing data
final_day_data = day_before_normalize
```

```python
#Features
X = final_day_data.drop(['cnt'], axis=1)

#Target variable
y = final_day_data['cnt']
```

### Regression Evaluation Metrics

1. Mean Absolute Error
2. Mean Squared Error
3. Root Mean Square Error

### We need to calculate the above metrics for other models too. So, we keep all these calculations in a function

```python
from sklearn import metrics
```

```python
def evaluate_metrics(y_test, predictions):
    print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, predictions))
    print('Mean Squared Error: ', metrics.mean_squared_error(y_test, predictions))
    print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
#Calculating Mean Absolute Percentage Error
def mape(true_target, prediction_target):
    result = np.mean(np.abs((true_target - prediction_target) / true_target)) * 100
    return result
```

```python
#Calling evaluate_metrics for Linear Regression
evaluate_metrics(y_test, predictions)
```

#### Mean Absolute Error            :  2.448
#### Mean Squared Error             :  8.866
#### Root Mean Squared Error  :  2.977

```python
#Calculating Mean Absolute Percentage Error
mape(y_test, predictions)
```

#### Mean Absolute Percentage Error = 0.0861
